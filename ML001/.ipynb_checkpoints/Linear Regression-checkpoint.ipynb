###### Importing Necessary Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###### Creating Regression Dataset

```python
from sklearn.datasets import make_regression
```

###### Creating X and y features

```python
X,y = make_regression(n_samples=10000,n_features=1,n_targets=1,bias=10,noise=5)
df = pd.DataFrame({"x":X.flatten(),'y':y.flatten()})
```

###### Testing Linearity

```python
sns.scatterplot(data=df,x='x',y='y')
plt.title("Scatterplot : Linearity")
plt.show()
```

```python
from scipy.stats import pearsonr
r,p = pearsonr(df['x'],df['y'])
if p >0.05:
    print("Their is no significant correlation")
else:
    print("Their is significant correlation")
```

###### Testing Normality for Dependent feature

```python
sns.kdeplot(data=df,x='x')
plt.title("KDE-plot : Normality")
plt.show()
```

```python
from scipy.stats import anderson,shapiro,kstest
print("Anderson Results")
s,cv,sl = anderson(df['y'])
results_a = pd.DataFrame(columns=['Statistic','Critical','SL','Status'])
results_a['Critical'] = cv
results_a["SL"] = sl
results_a['Statistic'] = s
results_a['Status'] = results_a["Statistic"]<results_a["Critical"]
results_a
```

```python
s,p = shapiro(df['y'])
if p >0.05:
    print("Normaly Distributed")
else:
    print("Not Normaly Distributed")
```

```python
s,p = kstest(df['y'],'norm')
if p >0.05:
    print("Normaly Distributed")
else:
    print("Not Normaly Distributed")
```

###### Simple Linear Regression Model using Statsmodel

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df['x'], df['y'], test_size=0.3, random_state=2806)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shap
```

```python
xtrain = xtrain.values.reshape(-1,1)
xtest = xtest.values.reshape(-1,1)
ytrain = ytrain.values.reshape(-1,1)
ytest = ytest.values.reshape(-1,1)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
xtrain = sm.add_constant(xtrain)
xtest = sm.add_constant(xtest)
```

```python
lm = sm.OLS(ytrain,xtrain).fit()
lm.summary()
```

```python
pred = lm.predict(xtest)
error = ytest.flatten()-pred
error
```

```python
sns.kdeplot(error);
```

```python
s,cv,sl = anderson(error)
results_a = pd.DataFrame(columns=['Statistic','Critical','SL','Status'])
results_a['Critical'] = cv
results_a["SL"] = sl
results_a['Statistic'] = s
results_a['Status'] = results_a["Statistic"]<results_a["Critical"]
results_a
```

###### Simple Linear Regression Model using Sklearn

```python
from sklearn.linear_model import LinearRegression,SGDRegressor
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df['x'], df['y'], test_size=0.3, random_state=2806)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
xtrain = xtrain.values.reshape(-1,1)
xtest = xtest.values.reshape(-1,1)
ytrain = ytrain.values.reshape(-1,1)
ytest = ytest.values.reshape(-1,1)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

```python
lm = LinearRegression()
lm.fit(xtrain,ytrain.flatten())
print(f"weights : {lm.coef_[0]}            Bias : {lm.intercept_}")
```

```python
sgd = SGDRegressor(max_iter=50,verbose=1)
sgd.fit(xtrain,ytrain.flatten())
print(f"weights : {sgd.coef_[0]}            Bias : {sgd.intercept_[0]}")
```

```python
pred = lm.predict(xtest)
error = ytest.flatten()-pred
error
```

```python
sns.kdeplot(error);
```

```python
s,cv,sl = anderson(error)
results_a = pd.DataFrame(columns=['Statistic','Critical','SL','Status'])
results_a['Critical'] = cv
results_a["SL"] = sl
results_a['Statistic'] = s
results_a['Status'] = results_a["Statistic"]<results_a["Critical"]
results_a
```
