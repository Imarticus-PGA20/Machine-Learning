## Importing Necessary Libraries

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###### Creating synthetic Datapoints

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=5000,n_features=1,n_targets=1,bias=10,noise=50)
df = pd.DataFrame({'x':x.flatten(),'y':y.flatten()})
df.head()
```

 - Checking for Linearity

```python
sns.scatterplot(data=df,x='x',y='y')
plt.title("Scatterplot : Linearity")
plt.show()
```

```python
sns.heatmap(data = df.corr(),annot=True,cmap='viridis');
```

```python
from scipy.stats import pearsonr
```

```python
r,p = pearsonr(df.x,df.y)
if p>0.05:
    print("Not Significant")
else:
    print("Significant")
```

- Normality of Dependent

```python
from scipy.stats import anderson, shapiro ,kstest
```

```python
sns.kdeplot(df.y);
```

```python
s,cv,sl = anderson(df.y)
a = pd.DataFrame(columns=['Statistic','Critical','Significance','Status'])
a['Critical'] = cv
a['Significance'] = sl
a['Statistic'] = s
a['Status'] = a["Statistic"] < a['Critical']
a
```

```python
s,p = shapiro(df.y)
if p>0.05:
    print("Significant")
else:
    print("Not Significant")
```

```python
s,p = kstest(df.y,'norm')
if p>0.05:
    print("Normal Distribution")
else:
    print("Not Normal")
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.x, df.y, test_size=0.3)
```

```python
xtrain = xtrain.values.reshape(-1,1)
xtest = xtest.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(xtrain,ytrain)
```

```python
pred = lm.predict(xtest)
pred
```

```python
ytest.values
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
np.sqrt(mean_squared_error(ytest,pred))
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
r2_score(ytest,pred)
```
